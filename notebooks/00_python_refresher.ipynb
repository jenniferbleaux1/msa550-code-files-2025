## Conditionals, Loops, Comprehensions


```python
# If/else conditional - only one branch of the code will run


age = 21

if age >= 18:
  print("You are an adult!")
else:
  print("You are a minor!")


 # Multi-branch conditions

score = 86

if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
elif score >= 60:
  print("D")
else:
  print("E")



# Loops - Looping with a range:  0, 1, 2


for i in range(3):  # range(0, 3, 1) --> range(start, end, increment)
  print(i)

# Looping over an item list

fruits = ["apple", "banana", "cherry"]

for f in fruits:
  print(f)

# List comprehension -

squares = [x**2 for x in range(10)]

# what is range(10)  --> 0, 1, 2, ..., 9
# x**2 for x         --> 0, 1, 4, ..., 81

print(squares)
```

## Collections you'll use most: lest and dict

```python
# Lists - ordered, changeable (mutable) sequences
numbers = [1, 2, 3, 4, 5]

print(numbers[2])   # get the number with index 2
numbers.append(6)   # add a number to the end
print(numbers)

# in excel you say A1:B5
print(numbers[:3])  #first three
print(numbers[-2:]) #last two
print(numbers[::2]) #start and end are default, step of two


# Dictionaries - key/value pair
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

print(person["name"])   # Lookup by key
person["age"] = 31      # Update a value using its key
person["country"] = "USA"
print(person)

# Tuples - immutable (unchangeable) sequence
number_tuple = (1, 2, 3, 4, 5)
print(number_tuple[0])  # Accessing elements by index

# Sets - unordered, unique
numbers_set = {1, 2, 3, 4, 5}
numbers_set.add(5)
print(numbers_set)

```

## NumPy Intro

## Pandas

```python
import pandas as pd
import numpy as np


data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago'],
    'Salary': [50000, 70000, 65000, 55000]
}


df = pd.DataFrame(data)
#print(df)
display(df)
```

```python
print("\nInfo:")
print(df.info())          # Schema/dtypes/non-null counts
print("\nDescribe:")
print(df.describe())      # Summary stats for numeric columns
```

```python
### Select, filter, transform
print("\nSelect columns:")
display(df[["Name","Salary"]])   ##  Project two columns by name

print("\nFilter (Age>30):")
display(df[df["Age"]>30])       # Row filter using a boolean condition

print("\nAdd column:")
df["Experience"] = [3, 7, 12, 5]  # Create a new column from a list (must match length)
display(df.head())

print("\nGroupby: mean salary by city")
display(df.groupby("City")["Salary"].mean())    # Aggregate after grouping by a categorical column
```

```python
### Missing values, dates, categoricals

# Introduce a missing value (NaN) â common in real datasets
df.loc[1, "Salary"] = np.nan
print("\nWith missing:")
display(df)

print("\nDrop missing rows:")
display(df.dropna())          # Remove rows with any NaNs (use cautiously)

print("\nFill with mean:")
display(df.fillna(df["Salary"].mean()))       # Imputation with the column mean (simple strategy)

# Create a DateTimeIndex and simple time series
dates = pd.date_range("2023-01-01", periods=6, freq="D")
ts_df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list("ABCD"))
print("\nTime series head:")
display(ts_df.head())

print("\nMonthly resample (mean):")
display(ts_df.resample("M").mean())     # Convert daily to monthly by averaging

# Convert City to a categorical dtype (saves memory; enables categorical ops)
df["City"] = df["City"].astype("category")
print("Categories:", df["City"].cat.categories)
```

```python
### Merge (join), reshape (melt/pivot)

# Build a small lookup table to join onto df
df2 = pd.DataFrame({
    "Name": ["Alice","Bob","Eve"],
    "Department": ["HR","IT","Finance"]
})

# Left join: keep all rows in df, bring Department where Name matches df2
merged = pd.merge(df, df2, on="Name", how="left")
print("\nMerged:")
display(merged)

# Reshape wide -> long (one row per metric per person)
long = pd.melt(merged, id_vars=["Name","City","Department"], value_vars=["Age","Salary","Experience"],
               var_name="metric", value_name="value")
print("\nLong format:")
display(long.head())

# Reshape long -> wide (opposite) using a pivot table
pivot = long.pivot_table(index="Name", columns="metric", values="value", aggfunc="mean")
print("\nBack to wide via pivot:")
display(pivot)
```

## Vizualization

```python
import numpy as np
import pandas as pd

# Keep plots reproducible
np.random.seed(42)


# Tiny synthetic dataset for visualization demos
people = ["Alice","Bob","Charlie","David","Eve","Frank","Grace","Heidi"]
age = np.random.randint(22, 55, size=len(people))
salary = np.random.randint(45_000, 120_000, size=len(people)).astype(float)
city = np.random.choice(["New York","San Francisco","Los Angeles","Chicago"], size=len(people))
experience = np.clip((age - 22) + np.random.randint(-2, 3, size=len(people)), 0, None)

df_viz = pd.DataFrame({
    "Name": people,
    "Age": age,
    "Salary": salary,
    "City": city,
    "Experience": experience
})
```

###1) Matplotlib â scatter + reference lines

Why this is useful: a simple, static plot for quick EDA. Adding mean reference lines and a small annotation helps students practice telling a data story (e.g., âwhoâs above/below average?â).

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))                            # Create a figure with a specific size
plt.scatter(df_viz["Age"], df_viz["Salary"], alpha=0.8)  # Scatter: Age vs Salary; alpha=transparency

plt.title("Salary vs Age (Matplotlib)")             # Title
plt.xlabel("Age")                                    # X-axis label
plt.ylabel("Salary")                                 # Y-axis label

# Reference lines: simple storytelling device for âabove/below averageâ
mu_age = df_viz["Age"].mean()
mu_sal = df_viz["Salary"].mean()
plt.axvline(mu_age, linestyle="--", linewidth=1)     # Vertical mean-age line
plt.axhline(mu_sal, linestyle="--", linewidth=1)     # Horizontal mean-salary line

# Minimal annotation to point out something interesting
plt.annotate(
    "Mean age",
    xy=(mu_age, df_viz["Salary"].min()),             # Arrow points to (x=mean age, y=min salary)
    xytext=(mu_age + 1, df_viz["Salary"].min() + 10_000),  # Place the text a bit to the right/up
    arrowprops=dict(arrowstyle="->")                  # Style of the arrow
)

plt.tight_layout()                                    # Adjusts paddings to avoid overlap
plt.show()                                            # Render the figure in the notebook
```

### 2) Seaborn â distribution comparison by category

Why this is useful: Seaborn makes statistical graphics easy. A boxplot shows spread/medians; the stripplot overlay reveals individual observations (great for small classes/data).

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()                                       # Pleasant defaults for style and color

plt.figure(figsize=(8,5))
sns.boxplot(data=df_viz, x="City", y="Salary")        # Boxplot: distribution per City
sns.stripplot(data=df_viz, x="City", y="Salary",      # Overlay raw points (great for small samples)
              alpha=0.7, dodge=True)

plt.title("Salary Distribution by City (Seaborn)")
plt.tight_layout()
plt.show()
```

### 3) Plotly â interactive scatter with hover

Why this is useful: Interactivity helps students explore: hover for details, zoom, and toggle series. Great for quick dashboards or in-class demos.

```python
import plotly.express as px

fig = px.scatter(
    df_viz,                               # DataFrame as the data source
    x="Age", y="Salary",                  # Axis mappings
    color="City",                         # Color groups (legend toggles)
    hover_data=["Name","Experience"],     # Extra fields shown in tooltip on hover
    title="Interactive Salary vs Age (Plotly)"  # Title displayed on the figure
)  # px.scatter returns a Figure object

fig.show()  # Render an interactive chart (zoom/pan, save, toggle series)
```

## Handy built-ins for analytics: math, statistics, random


```python
import math, statistics, random

data = [2,2,3,4,4,4,5,5,5,5,6,6,7]                         # A small numeric list

# A few summary statistics from the 'statistics' module
print("Mean/Median/Mode:", statistics.mean(data),          # Arithmetic mean
      statistics.median(data),                             # Middle value
      statistics.mode(data))                               # Most frequent value

print("Variance/Stdev:", statistics.variance(data),        # Sample variance
      statistics.stdev(data))                              # Sample standard deviation

# Normal distribution helper: cumulative probability at 1.96 (~97.5%)
print("Normal cdf at 1.96:", statistics.NormalDist(0,1).cdf(1.96))

# Random module: sampling & shuffling
print("Random int 1-10:", random.randint(1,10))            # Inclusive endpoints
items = ['apple','banana','cherry','date']
random.shuffle(items)                                       # In-place reordering
print("Shuffled:", items[:3])                               # Show first three items
```
